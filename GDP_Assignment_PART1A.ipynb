#Calculate the average growth of states for the duration 2013-14, 2014-15 and 2015-16 by taking the mean of the row '(% Growth over previous year)

```python
df2 = df.iloc[[7,8,9]]
df2
```

```python
df3 = df.iloc[[7,8,9]].mean(axis=0)
df3.describe()
df3
```

```python
barwidth = 0.25
df2.groupby(['Duration']).plot(kind='bar' ,figsize=(20,10))
plt.xlabel("State")
plt.ylabel("Avg.%age growth over Duration")
plt.title('Each State Avg %age growth over the Duration')
```

```python
df4 = df3.rename_axis("STATE", axis=0)
df4
```

```python
# set a seaborn style of your taste
sns.set(style="white", rc={"lines.linewidth": 1})
plt.figure(num=None, figsize=(12, 8), dpi=80, facecolor='w', edgecolor='k')
sns.set(rc={'figure.figsize':(20,15)})
sns.barplot(df4.index,df4.values)
plt.xlabel("State")
plt.ylabel("Avg.%age growth over previous year")
plt.title('Each State Avg %age growth over the previous year 2013-2016')
plt.show()
```

Another and clear way of representation:-

```python
plt.xlabel("State")
plt.ylabel("Avg.%age growth over previous year")
plt.title('Each State Avg %age growth over the previous year 2013-2016')
df4.plot.bar()
```
